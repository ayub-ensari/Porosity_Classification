```python
import cv2
import os
import glob
import numpy as np
from os import walk
import time
from matplotlib import pyplot as plt
import matplotlib
import pandas as pd
import tensorflow as tf
```

```python
def load_all_image_path(img_dir):
    
    #img_dir = "./Build2" # Enter Directory of all images
    img_labels = []
    for(_, _, filenames) in walk(img_dir):
        img_labels.extend(filenames)
        break
    data_path = os.path.join(img_dir,'*g')
    files = glob.glob(data_path)
    return files, img_labels
    
```

### Layers with porosity

```python
b1_prosity_index = list(range(311,378)) + list(range(441, 448)) + list(range(536, 553)) + list(range(627, 662)) + list(range(827, 862)) + list(range(936, 953)) + list(range(1011, 1078)) + list(range(1136, 1153)) 
b2_prosity_index = list(range(311,378)) + list(range(427, 462)) + list(range(531, 558)) + list(range(636, 653)) + list(range(737, 752)) + list(range(841, 848))
b3_prosity_index = list(range(402,437)) + list(range(506, 533)) + list(range(611, 628)) + list(range(712, 727)) + list(range(816, 823)) + list(range(917, 922))
```

```python
def crop_save_images(files, directory, labels):
    b1_prosity_index = list(range(311,378)) + list(range(441, 448)) + list(range(536, 553)) + list(range(627, 662)) + list(range(827, 862)) + list(range(936, 953)) + list(range(1011, 1078)) + list(range(1136, 1153)) 
    b2_prosity_index = list(range(311,378)) + list(range(427, 462)) + list(range(531, 558)) + list(range(636, 653)) + list(range(737, 752)) + list(range(841, 848))
    b3_prosity_index = list(range(402,437)) + list(range(506, 533)) + list(range(611, 628)) + list(range(712, 727)) + list(range(816, 823)) + list(range(917, 922))
    
    for f1,lab in zip(files,labels):
        #print("F1: " + str(f1))
        #print("Lab: " +  str(lab))
        ########## read image
        orig_img = cv2.imread(f1)

        ########### crop image
        img = orig_img[1250:1440, 650:1100]
        img1 = img[0:190,0:150]
        img2 = img[0:190,150:300]
        img3 = img[0:190,300:450]

        ########### Label Image
        
        tt = lab[:-4].split('_')
        print(tt[3])
        layer_no = int(tt[3])
        
        if (layer_no in b1_prosity_index):
            img_name_b1 = "1_B1_Layer_"+str(layer_no)+".jpg"
            #print(layer_no)
            #print("True")
        else:
            img_name_b1 = "0_B1_Layer_"+str(layer_no)+".jpg"
            #print("False")
            
        if (layer_no in b2_prosity_index):
            img_name_b2 = "1_B2_Layer_"+str(layer_no)+".jpg"
            #print(layer_no)
            #print("True")
        else:
            img_name_b2 = "0_B2_Layer_"+str(layer_no)+".jpg"
            #print("False")
            
        if (layer_no in b3_prosity_index):
            img_name_b3 = "1_B3_Layer_"+str(layer_no)+".jpg"
            #print(layer_no)
            #print("True")
        else:
            img_name_b3 = "0_B3_Layer_"+str(layer_no)+".jpg"
            #print("False")
        ########### store image
        if(layer_no>243 and layer_no<1243):
            img_name = directory[0] + img_name_b1
            matplotlib.image.imsave(img_name, img1)
        
            img_name = directory[1] + img_name_b2
            matplotlib.image.imsave(img_name, img2)
        if(layer_no>218 and layer_no<1218):
            img_name = directory[2] + img_name_b3
            matplotlib.image.imsave(img_name, img3)
        #break
```

```python
directories = ["./Crop_images/B1/", "./Crop_images/B2/", "./Crop_images/B3/"]
crop_save_images(img_paths[217:1206] ,directories, labels[217:1206]) 
```

##  Loading cropped images

```python
files, labels = load_all_image_path("./Crop_images/B1")
data = []
b1_labels = list()
for f1, lab in zip(files, labels):
    b1_labels.append(int(lab[0]))
    img = cv2.imread(f1)
    data.append(img)
b1_images = np.array(data)
```

```python
files, labels = load_all_image_path("./Crop_images/B2")
data = []
b2_labels = list()
for f1, lab in zip(files, labels):
    img = cv2.imread(f1)
    b2_labels.append(int(lab[0]))
    data.append(img)
b2_images = np.array(data)
```

```python
(unique, counts) = np.unique(b2_labels, return_counts=True)
print(unique, counts)
print(b2_images.shape)
```

```python
files, labels = load_all_image_path("./Crop_images/B3")
data = []
b3_labels = list()
for f1, lab in zip(files, labels):
    img = cv2.imread(f1)
    b3_labels.append(int(lab[0]))
    data.append(img)
b3_images = np.array(data)
```

```python
def visualize_n_imgs(PBI_data,labels, n):
    
    rnd_indx = np.random.choice(PBI_data.shape[0], n, replace=False)
    print(rnd_indx)
    data = []
    titles = []
    for i in range(0,n):
        index = rnd_indx[i]
        titles.append(labels[index])
        img = PBI_data[index]#:,:,:]
        data.append(img)
    test_images = np.array(data)

    plt.figure(figsize=(20,10))
    for i in range(n):
        plt.subplot(n/4,4,i+1),plt.imshow(test_images[i],'gray')
        plt.title(titles[i])
        plt.xticks([]),plt.yticks([])

    plt.show()
```

```python
visualize_n_imgs(train_images,train_labels, 12)
```

### Data Pre processing

```python
# Convert the array to float32 as opposed to uint8
train_images = train_images.astype(np.float32)
test_images = test_images.astype(np.float32)

# Convert the pixel values from integers between 0 and 255 to floats between 0 and 1
train_images /= 255
test_images /=  255
```

```python
NUM_DIGITS = 2

print("Before", b1_labels[311]) # The format of the labels before conversion

train_labels  = tf.keras.utils.to_categorical(train_labels, NUM_DIGITS)

print("After", b1_labels[311]) # The format of the labels after conversion

test_labels = tf.keras.utils.to_categorical(test_labels, NUM_DIGITS)
```

### Model

```python
model = tf.keras.Sequential()

model.add(tf.keras.layers.Conv2D(filters=32, kernel_size=(5, 5), padding='same', activation='relu', input_shape=(190, 150,3)))
model.add(tf.keras.layers.MaxPool2D(strides=2))
model.add(tf.keras.layers.Conv2D(filters=48, kernel_size=(5, 5), padding='valid', activation='relu'))
model.add(tf.keras.layers.MaxPool2D(strides=2))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(512, activation=tf.nn.relu, input_shape=(85500,)))
model.add(tf.keras.layers.Dense(256, activation='relu'))
model.add(tf.keras.layers.Dense(84, activation='relu'))
#model.add(tf.keras.layers.Dense(10, activation='softmax'))



model.add(tf.keras.layers.Dense(2, activation=tf.nn.softmax))

# We will now compile and print out a summary of our model
model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

model.summary()
```

```python
loss, accuracy = model.evaluate(test_images, test_labels)
print('Test accuracy: %.2f' % (accuracy))
```
